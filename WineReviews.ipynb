```python
#  Read in data

reviews = pd.read_csv("data/winemag-data-130k-v2.csv.zip", index_col=0)

#  Create a summary of the data that contains the name, number of reviews, 
#  and the average points for each unique country in the dataset

country_Name = reviews.country.unique ()
review_Num = reviews.country.value_counts()
avg_Points = reviews.groupby('country').points.mean().round(1)
wineReviews = pd.DataFrame({'country':country_Name, 
                            'count' :review_Num.reindex(country_Name),
                            'points' :avg_Points.reindex(country_Name)})

# I'm not sure how to remove the duplicate country column. 

# Method 2 that errored out when running:
# wineReviews = reviews.groupby('country').agg(
#     number_of_reviews=('country', 'size'),
#     average_points=('points', 'mean')
# )
# wineReviews.columns = ['country', 'number_of_reviews', 'average_points']


wineReviews

```
